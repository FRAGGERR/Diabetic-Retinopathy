ACCURACY PLOT

```python
import matplotlib.pyplot as plt
import numpy as np

# Replace these lists with the actual metrics from your training
metrics = ['DenseNet121', 'DenseNet169', 'DenseNet201', 'InceptionV3', 'MobileNet', 'ResNet152', 'Xception']
train_values = [77.41, 89.68, 88.86, 92.82, 59.86, 60.24, 89.53]
val_values = [72.55, 74.36, 74.55, 74.50, 62.55, 63.18, 73.10]

# Number of metrics
num_metrics = len(metrics)

# Plotting with larger figure size and adjusted bar width
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust the figure size as needed
bar_width = 0.35  # Adjust the bar width as needed
index = np.arange(num_metrics)

train_bars = ax.bar(index, train_values, bar_width, label='Training', color='#f0ead2', edgecolor='black')
val_bars = ax.bar(index + bar_width, val_values, bar_width, label='Validation', color='#adc178', edgecolor='black')

# Adding labels and title
ax.set_xlabel('Accuracy')
ax.set_ylabel('Percentage')
ax.set_title('All Models Accuracy Plot')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(metrics)
ax.legend()
# Set y-axis range to 0 to 120
ax.set_ylim(0, 120)

# Set y-axis ticks and labels up to 100
ax.set_yticks(range(0, 101, 20))
ax.set_yticklabels(range(0, 101, 20))

# Adding values on top of the bars with adjusted vertical offset
def add_values(bars, values, offset):
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax.annotate('{}'.format(value),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, offset),  # Adjusted vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Adding values for training bars with a smaller vertical offset
add_values(train_bars, train_values, 1)
# Adding values for validation bars with a smaller vertical offset
add_values(val_bars, val_values, 1)

plt.tight_layout()  # Ensure labels and ticks are not cut off
plt.show()
```

ALL TOGETHER

```python
import matplotlib.pyplot as plt
import numpy as np

# Model names
models = ['DenseNet121', 'DenseNet169', 'DenseNet201', 'InceptionV3', 'MobileNet', 'ResNet152', 'Xception']

# Accuracy values
train_accuracy = [77.41, 89.68, 88.86, 92.82, 59.86, 60.24, 89.53]
val_accuracy = [72.55, 74.36, 74.55, 74.50, 62.55, 63.18, 73.10]

# Precision values
train_precision = [87.78, 91.98, 92.70, 94.76, 91.40, 89.24, 91.58]
val_precision = [85.14, 79.05, 83.27, 76.47, 95.92, 95.87, 78.37]

# Recall values
train_recall = [74.12, 88.36, 86.68, 91.34, 38.62, 41.12, 87.01]
val_recall = [71.35, 72.74, 73.09, 71.06, 57.18, 55.90, 71.46]

# F1 Score values
train_f1 = [76.84, 90.10, 89.53, 92.99, 50.73, 60.12, 89.18]
val_f1 = [73.11, 74.98, 75.06, 73.20,60.48, 53.06, 72.89]

# Number of models
num_models = len(models)

# Plotting
bar_width = 0.105
index = np.arange(num_models)

fig, ax = plt.subplots(figsize=(20, 15))  # Increase figure size

# Function to add text values
def add_values(bars, values, offset):
    for bar, value in zip(bars, values):
        width = bar.get_width()
        ax.text(width + offset, bar.get_y() + bar.get_height() / 2, '{:.2f}'.format(value),
                ha='left', va='center', color='black')

# Grouped horizontal bar chart for Accuracy
train_accuracy_bars = ax.barh(index - bar_width, train_accuracy, bar_width, label='Train Accuracy', color='royalblue', edgecolor='black')
val_accuracy_bars = ax.barh(index, val_accuracy, bar_width, label='Validation Accuracy', color='darkorange', edgecolor='black')
add_values(train_accuracy_bars, train_accuracy, 2)
add_values(val_accuracy_bars, val_accuracy, 2)

# Grouped horizontal bar chart for Precision
train_precision_bars = ax.barh(index + bar_width, train_precision, bar_width, label='Train Precision', color='mediumseagreen', edgecolor='black')
val_precision_bars = ax.barh(index + 2 * bar_width, val_precision, bar_width, label='Validation Precision', color='tomato', edgecolor='black')
add_values(train_precision_bars, train_precision, 2)
add_values(val_precision_bars, val_precision, 2)

# Grouped horizontal bar chart for Recall
train_recall_bars = ax.barh(index + 3 * bar_width, train_recall, bar_width, label='Train Recall', color='mediumpurple', edgecolor='black')
val_recall_bars = ax.barh(index + 4 * bar_width, val_recall, bar_width, label='Validation Recall', color='gold', edgecolor='black')
add_values(train_recall_bars, train_recall, 2)
add_values(val_recall_bars, val_recall, 2)

# Grouped horizontal bar chart for F1 Score
train_f1_bars = ax.barh(index + 5 * bar_width, train_f1, bar_width, label='Train F1 Score', color='lightcoral', edgecolor='black')
val_f1_bars = ax.barh(index + 6 * bar_width, val_f1, bar_width, label='Validation F1 Score', color='lightseagreen', edgecolor='black')
add_values(train_f1_bars, train_f1, 2)
add_values(val_f1_bars, val_f1, 2)

# Adding labels and title
ax.set_ylabel('Models')
ax.set_xlabel('Percentage')
ax.set_title('Performance Metrics - Train and Validation Comparison')
ax.set_yticks(index + 2 * bar_width)
ax.set_yticklabels(models)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Set x-axis range to 0 to 120
ax.set_xlim(0, 120)

# Set x-axis ticks and labels up to 100
ax.set_xticks(range(0, 101, 20))
ax.set_xticklabels(range(0, 101, 20))

plt.tight_layout()
plt.show()
```
